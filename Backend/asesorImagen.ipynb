```python
import mediapipe as mp
from mediapipe.tasks import python
from mediapipe.tasks.python import vision
import numpy as np
import cv2
```

# Con imágenes

```python
from mediapipe.framework.formats import landmark_pb2
import cv2
# Load the input image from an image file.

model_path="pose_landmarker_heavy.task"
BaseOptions = mp.tasks.BaseOptions
PoseLandmarker = mp.tasks.vision.PoseLandmarker
PoseLandmarkerOptions = mp.tasks.vision.PoseLandmarkerOptions
VisionRunningMode = mp.tasks.vision.RunningMode


options = PoseLandmarkerOptions(
    base_options=BaseOptions(model_asset_path=model_path),
    running_mode=VisionRunningMode.IMAGE)

mp_drawing = mp.solutions.drawing_utils
mp_pose = mp.solutions.pose


with PoseLandmarker.create_from_options(options) as landmarker:
    mp_image = mp.Image.create_from_file('imagenPrueba1.jpg')
    imagen_copy=mp_image.numpy_view().copy()
    pose_landmarker_result = landmarker.detect(mp_image)
    print(pose_landmarker_result)
    for normalized_landmarks in pose_landmarker_result.pose_landmarks:
        landmark_list = landmark_pb2.NormalizedLandmarkList()
        landmark_list.landmark.extend([
            landmark_pb2.NormalizedLandmark(
                x=lm.x, y=lm.y, z=lm.z) for lm in normalized_landmarks
        ])
        
        mp_drawing.draw_landmarks(
            imagen_copy,
            landmark_list, 
            mp_pose.POSE_CONNECTIONS,
            mp_drawing.DrawingSpec(color=(245, 117, 66), thickness=2, circle_radius=2),
            mp_drawing.DrawingSpec(color=(245, 66, 230), thickness=2)
        )
    imagen_final = cv2.cvtColor(imagen_copy, cv2.COLOR_RGB2BGR) 
    #Guardamos lo que detecta mediapipe pose
    cv2.imwrite('imagen_con_pose.jpg', imagen_final)
    print("✅ Imagen guardada como 'imagen_con_pose.jpg'")
   
    
    
 
    

```

```python
#Visualiza resultados 
import cv2
imagen_final = cv2.imread('imagen_con_pose.jpg')
cv2.imshow('Pose Landmarker', imagen_final)
if cv2.waitKey(0) & 0xFF == 27:
    cv2.destroyAllWindows()
```
